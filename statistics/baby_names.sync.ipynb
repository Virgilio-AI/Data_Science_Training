```python
# The data set contains the top 1000 baby names in the US
# for each year from 1880 until 2008
names_df = pd.read_csv('data/baby-names2.csv')
names_df.head()
```

```python
boys_names = names_df[names_df['sex'] == 'boy']
```

```python

def get_max_proportion(group):
	return group.loc[group['prop'].idxmax()]
```

```python
max_prop_boys = boys_names.groupby('year').apply(get_max_proportion)
```

```python
max_prop = names_df.groupby('year').apply(get_max_proportion)
```
